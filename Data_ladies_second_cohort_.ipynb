## Creating Arrays

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0])
print(arr[1:4])
print(arr[-1])
```

## Array Shape & Size

```python
arr2d = np.array([[1, 2, 3], [4, 5, 6]])


print("Shape:", arr2d.shape)
print("Size:", arr2d.size)
print("Rows:", arr2d[0])
print("Specific Element (row 1, col 2):", arr2d[1, 2])
```

## Math Operations

```python
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])


print(a + b)
print(a * b)
print(b - a)
print(b / a)
```

## Statistics with NumPy

```python
stats_arr = np.array([4, 6, 8, 10, 12])


print("Mean:", stats_arr.mean())
print("Median:", np.median(stats_arr))
print("Min:", stats_arr.min())
print("Max:", stats_arr.max())
print("Std Dev:", stats_arr.std())
```

# Loading & Exploring Data

```python
df = pd.read_csv("sample.csv")


print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

## Selecting Columns & Rows

```python
# Selecting columns
ages = df["Age"]


# Selecting multiple columns
subset = df[["Name", "Score"]]


# Selecting rows by index
first_row = df.iloc[0]
rows_1_to_3 = df.iloc[1:4]


ages, subset, first_row, rows_1_to_3
```

## Sorting Data

```python
# Sort by age
sorted_df = df.sort_values(by="Age", ascending=True)
sorted_df
```

## Handling Missing Values

```python
# Check for missing values
print(df.isnull().sum())


# Fill missing values
filled_df = df.fillna({"Age": df["Age"].mean()})


# Drop rows with missing values
dropped_df = df.dropna()
```

## Removing Duplicates

```python
# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()


# Count duplicates
print("Duplicates removed:", len(df) - len(df_no_duplicates))
```

## Renaming Columns




```python
# Rename specific columns
df_renamed = df.rename(columns={"Name": "Full_Name", "Score": "Exam_Score"})
df_renamed.head()
```

## Changing Data Types

```python
# Show current types
print(df.dtypes)


# Convert Age to float
df["Age"] = df["Age"].astype(float)
print(df.dtypes)
```

## Replacing Incorrect Data

```python
# Example: Replace wrong age values
df["Age"] = df["Age"].replace({0: None})
df["Age"] = df["Age"].fillna(df["Age"].mean())


df.head()
```

## Data Analysis

```python
# Students older than 24
older_students = df[df["Age"] > 24]


# Score above 85
high_scores = df[df["Score"] > 85]


older_students, high_scores
```

## Grouping & Aggregation

```python
# Example dataset with categories


sample = pd.DataFrame({
"Department": ["IT", "HR", "IT", "Finance", "HR"],
"Salary": [800, 600, 900, 700, 650]
})


# Group by department
print(sample.groupby("Department")["Salary"].mean())
```

## Sorting & Ranking

```python
# Sort by scores in descending order
sorted_scores = df.sort_values(by="Score", ascending=False)


# Rank scores
ranked_df = df.copy()
ranked_df["Rank"] = df["Score"].rank(ascending=False)


sorted_scores, ranked_df
```
